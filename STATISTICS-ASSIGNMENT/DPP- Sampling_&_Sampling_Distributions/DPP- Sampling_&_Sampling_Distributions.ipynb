## Question 1: Suppose you want to study customer satisfaction in a mall. Explain how you would use systematic sampling to select participants. 

**Definitions:**  
- `N` = total number of customers  
- `n` = required sample size  
- `k` = sampling interval

**Formula:**

$$ k = \frac{N}{n} $$

**Steps:**
1. Arrange customers in the order of entry.
2. Choose a random starting point between 1 and `k`.
3. Select every `k`-th customer thereafter.

This method ensures uniform coverage and simplicity.

## Question 2: A school has 5 grades, each with different numbers of students. Describe how you would use stratified sampling to estimate the average exam score of all students. 

The population is divided into **homogeneous subgroups (strata)** based on grades.

**Formula for overall mean:**

$$ \bar{X} = \sum (w_i \times \bar{X}_i) $$

where:  
- $w_i$ = proportion of students in grade *i*  
- $\bar{X}_i$ = mean score of grade *i*

This guarantees representation from all grades.

## Question 3: A school has 800 students: 400 in junior grades, 250 in senior grades, and 150 in advanced grades. You want to survey 100 students ensuring proportional representation. 

**Given:**  
- Total students, $N = 800$  
- Sample size, $n = 100$

**Formula:**

$$ n_i = \frac{N_i}{N} \times n $$

**Calculations:**

- Junior grades: $(400/800) \times 100 = 50$  
- Senior grades: $(250/800) \times 100 \approx 31$  
- Advanced grades: $(150/800) \times 100 \approx 19$

## Question 6: In a population of students, the average study time is 10 hours per week with a standard deviation of 3 hours. A sample of size 36 is taken. 

**Given:**  
- Population standard deviation, $\sigma = 3$  
- Sample size, $n = 36$

**Formula:**

$$ SE = \frac{\sigma}{\sqrt{n}} $$

**Calculation:**

$$ SE = \frac{3}{\sqrt{36}} = \frac{3}{6} = 0.5 $$

## Question 7: If factory light bulb lifespans have a mean of 1,200 hours and a standard deviation of 200 hours, calculate the standard error for samples of size 50. 

**Given:**  
- $\sigma = 200$, $n = 50$

**Formula:**

$$ SE = \frac{\sigma}{\sqrt{n}} $$

**Calculation:**

$$ SE = \frac{200}{\sqrt{50}} \approx 28.28 $$

## Question 8: A population has a mean of 50 and a standard deviation of 10. Compare the standard error for sample sizes of 16, 64, and 256. 

**Formula:**

$$ SE = \frac{\sigma}{\sqrt{n}} $$

**Results:**

- $n = 16 \Rightarrow SE = 2.5$  
- $n = 64 \Rightarrow SE = 1.25$  
- $n = 256 \Rightarrow SE = 0.625$

**Conclusion:** As sample size increases, standard error decreases.

## Question 11: The average weight of apples is 200 g with a standard deviation of 50 g. A random sample of 36 apples is taken. What is the probability that the sample mean is greater than 210 g? 

**Given:**  
- Population mean, $\mu = 200$  
- Population standard deviation, $\sigma = 50$  
- Sample size, $n = 36$

**Step 1: Standard Error**

$$ SE = \frac{50}{\sqrt{36}} = 8.33 $$

**Step 2: Z-score**

$$ Z = \frac{210 - 200}{8.33} = 1.20 $$

**Step 3: Probability**

From Z-table:

$$ P(Z < 1.20) = 0.8849 $$

$$ P(\bar{X} > 210) = 1 - 0.8849 = 0.115 $$

Thus, there is an 11.5% probability that the sample mean exceeds 210 g.
